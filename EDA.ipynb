# RFM Analysis

```python
df
```

```python
# Ensure the date column is in datetime format
df['ORDER_CREATED_DATE'] = pd.to_datetime(df['ORDER_CREATED_DATE'])

# Set a "snapshot date" to calculate recency from. This is the day after the last order.
snapshot_date = df['ORDER_CREATED_DATE'].max() + dt.timedelta(days=1)

# Group by customer and calculate R, F, and M values
rfm_data = df.groupby('CUSTOMER_ID').agg({
    'ORDER_CREATED_DATE': lambda date: (snapshot_date - date.max()).days, # Recency
    'ORDER_ID': 'nunique',                                                # Frequency
    'total_sales': 'sum'                                                  # Monetary
})

# Rename the columns for clarity
rfm_data.rename(columns={
    'ORDER_CREATED_DATE': 'Recency',
    'ORDER_ID': 'Frequency',
    'total_sales': 'Monetary'
}, inplace=True)

print("Raw RFM Values:")
rfm_data
```

```python
# Create labels for scoring (5 is best)
r_labels = range(5, 0, -1) # Lower Recency (fewer days) gets a higher score
f_labels = range(1, 6)     # Higher Frequency gets a higher score
m_labels = range(1, 6)     # Higher Monetary value gets a higher score

# Assign scores to each customer based on quintiles
rfm_data['R_score'] = pd.qcut(rfm_data['Recency'], q=5, labels=r_labels, duplicates='drop').astype(int)
rfm_data['F_score'] = pd.qcut(rfm_data['Frequency'].rank(method='first'), q=5, labels=f_labels).astype(int)
rfm_data['M_score'] = pd.qcut(rfm_data['Monetary'], q=5, labels=m_labels, duplicates='drop').astype(int)

print("\nRFM Scores:")
rfm_data
```

```python
# Combine R, F, and M scores into a single RFM_Score
rfm_data['RFM_Score'] = rfm_data['R_score'].astype(str) + rfm_data['F_score'].astype(str) + rfm_data['M_score'].astype(str)

# Define segment mapping based on score patterns
segment_map = {
    r'[4-5][4-5][4-5]': 'Champions',
    r'[3-5][3-5][1-3]': 'Loyal Customers',
    r'[4-5]1[1-5]': 'New Customers',
    r'3[1-3][1-3]': 'Potential Loyalists',
    r'[1-2][3-5][3-5]': 'At-Risk Customers',
    r'[1-2]1[1-2]': 'Lost Customers'
}

# Assign segment names using the map
rfm_data['Segment'] = rfm_data['RFM_Score'].replace(segment_map, regex=True)

# Fill any unmapped scores with a default value
rfm_data['Segment'] = rfm_data['Segment'].where(rfm_data['Segment'].isin(segment_map.values()), 'Other')

print("\nFinal Customer Segments:")
rfm_data
```

```python
# Group by segment, sum sales, and format the output
sales_by_segment = (x.groupby(by='Segment')['total_sales']
                    .sum()
                    .div(1_000_000)  # Divide by 1 million
                    .round(2)       # Round to 2 decimal places
                    .reset_index()  # Convert to a DataFrame
                    .rename(columns={'total_sales': 'Total Sales (in Millions)'}) # Rename column for clarity
                    .sort_values(by='Total Sales (in Millions)', ascending=False)) # Sort by the new column

sales_by_segment
```

```python
# First, calculate the total sales for each unique order, keeping the segment information
order_totals_by_segment = x.groupby(['Segment', 'ORDER_ID'])['total_sales'].sum()

# Then, calculate the average of those order totals for each segment
aov_by_segment = (order_totals_by_segment.groupby('Segment')
                  .mean()
                  .round(2)
                  .reset_index()
                  .sort_values(by='total_sales', ascending=False)
                  .rename(columns={'total_sales': 'AOV'})) # Rename for clarity

aov_by_segment
```
