# 1. Registered and eClub

```python
df1 = df_edible_cust3[df_edible_cust3['CUSTOMER_TYPE'] != 'Guest']
df1
```

```python
df1.nunique()
```

```python
# Create the pivot table from your loyal customer dataframe
user_item_matrix_loyal = df1.pivot_table(
    index='CUSTOMER_ID',
    columns='item_name',
    values='item_quantity',
    aggfunc='sum',
    fill_value=0
)

user_item_matrix_loyal
```

```python
from scipy.sparse import csr_matrix
from sklearn.metrics.pairwise import cosine_similarity

# Convert to a sparse matrix for efficiency
user_item_sparse_loyal = csr_matrix(user_item_matrix_loyal.values)

# Compute cosine similarity between items
item_similarity_loyal = cosine_similarity(user_item_sparse_loyal.T)

# Create the final item similarity dataframe for easy lookup
item_similarity_df_loyal = pd.DataFrame(item_similarity_loyal, index=user_item_matrix_loyal.columns, columns=user_item_matrix_loyal.columns)

print("âœ… Collaborative Filtering model for loyal customers is trained.")

item_similarity_df_loyal
```

```python
def get_personalized_recommendations(cart_items, similarity_df):
    """Generates personalized recommendations using an item similarity matrix."""
    
    similar_scores = pd.Series(dtype=float)

    # For each item in the cart, get its similarity scores with all other items
    for item in cart_items:
        if item in similarity_df.columns:
            # Use concat to append the series of scores for the current item
            similar_scores = pd.concat([similar_scores, similarity_df[item]])
    
    # If no items in the cart are in our model, we can't recommend anything
    if similar_scores.empty:
        return ["N/A", "N/A", "N/A"]

    # Group by item name and take the mean similarity score
    grouped_scores = similar_scores.groupby(similar_scores.index).mean()

    # Sort the items by their average similarity score in descending order
    sorted_scores = grouped_scores.sort_values(ascending=False)

    # Filter out any items that are already in the cart
    recommendations = [item for item in sorted_scores.index if item not in cart_items]
    
    # Pad with placeholders if we have fewer than 3 recommendations
    while len(recommendations) < 3:
        recommendations.append("N/A")
        
    return recommendations[:3]
```

```python
# Example cart for a loyal customer
example_cart = {'6pc Boneless Mild', 'Large Cheese Fries', '20 Oz Soda'}

# Call the personalized recommendation function
personalized_recs = get_personalized_recommendations(example_cart, item_similarity_df_loyal)

print(f"For the cart: {example_cart}")
print(f"Personalized recommendations are: {personalized_recs}")
```

# 2. Guest

```python
df2 = df_edible_cust3[df_edible_cust3['CUSTOMER_TYPE'] == 'Guest']
df2
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth, association_rules

# Create transactions from the guest dataframe
transactions_guest = df2.groupby('ORDER_ID')['item_name'].apply(list).tolist()

# One-hot encode the guest transactions
te_guest = TransactionEncoder()
te_ary_guest = te_guest.fit(transactions_guest).transform(transactions_guest)
market_basket_guest = pd.DataFrame(te_ary_guest, columns=te_guest.columns_)

market_basket_guest
```

```python
# Find frequent itemsets for guests
frequent_itemsets_guest = fpgrowth(market_basket_guest, min_support=0.001, use_colnames=True)

# Generate the final association rules for guests
rules_guest = association_rules(frequent_itemsets_guest, metric="lift", min_threshold=1)

print("âœ… FP-Growth model for Guest customers is trained.")

rules_guest
```

```python
def get_guest_recommendations(cart_items, rules_df):
    """Generates non-personalized recommendations using association rules."""
    
    # Find rules where the 'antecedents' are a subset of the items in the cart
    matching_rules = rules_df[rules_df['antecedents'].apply(lambda x: x.issubset(cart_items))]

    if matching_rules.empty:
        return ["N/A", "N/A", "N/A"]

    # Sort the matching rules by 'lift' to prioritize the strongest associations
    sorted_rules = matching_rules.sort_values(by='lift', ascending=False)
    
    # Extract the recommended items (consequents)
    recommendations = sorted_rules['consequents'].explode().unique()
    
    # Filter out items that are already in the cart
    final_recs = [rec for rec in recommendations if rec not in cart_items]
    
    # Pad with placeholders if we have fewer than 3 recommendations
    while len(final_recs) < 3:
        final_recs.append("N/A")

    return final_recs[:3]
```

```python
# Example cart for a guest customer
example_guest_cart = {'6pc Boneless Mild', 'Large Cheese Fries', '20 Oz Soda'}

# Call the guest recommendation function with the guest rules
guest_recs = get_guest_recommendations(example_guest_cart, rules_guest)

print(f"For the guest cart: {example_guest_cart}")
print(f"Guest recommendations are: {guest_recs}")
```

# Manager Recommendation Function

```python
def get_final_recommendations(cart_items, customer_type, rules_guest, item_similarity_df_loyal):
    """
    This function acts as a manager, choosing the correct model based on customer type.
    """
    
    # If the customer is a Guest, use the FP-Growth model
    if customer_type == 'Guest':
        return get_guest_recommendations(cart_items, rules_guest)
        
    # If the customer is Registered or eClub, use the personalized Collaborative Filtering model
    else: # This covers 'Registered' and 'eClub'
        return get_personalized_recommendations(cart_items, item_similarity_df_loyal)
```

```python
# A list to store the final recommendations for each order
recommendations_list = []

# Loop through each row of the test_data
for index, row in test_data.iterrows():
    # Create the cart from the item columns
    cart = {row['item1'], row['item2'], row['item3']}
    
    # Get the customer type for this order
    cust_type = row['CUSTOMER_TYPE']
    
    # Call the single manager function with all the necessary info
    recommendations = get_final_recommendations(
        cart, 
        cust_type, 
        rules_guest,                 # Pass in the trained guest model
        item_similarity_df_loyal     # Pass in the trained loyal customer model
    )
    
    recommendations_list.append(recommendations)

# --- Create the final submission DataFrame ---
# Convert the list of recommendations into a DataFrame
recommendations_df = pd.DataFrame(recommendations_list, columns=['RECOMMENDATION 1', 'RECOMMENDATION 2', 'RECOMMENDATION 3'])

# Assign the new columns to your original test_data
test_data[['RECOMMENDATION 1', 'RECOMMENDATION 2', 'RECOMMENDATION 3']] = recommendations_df

print("âœ… Recommendations have been successfully added to the test data.")
test_data
```
